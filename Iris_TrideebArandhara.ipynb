```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.tail()
```

```python
iris.info()
```

```python
iris.describe()
```

```python
np.round(iris.describe(),2)
```

```python
np.round(iris.select_dtypes(include='number').mean(), 2)
```

```python
np.round(iris.select_dtypes(include='number').median(), 2)
```

```python
iris.count()
```

```python
pd.crosstab(index=iris['species'], columns='frequency')
```

```python
correlaton = iris.corr(numeric_only=True)
correlaton
```

```python
sns.heatmap(correlaton, annot=True, cmap='coolwarm')
plt.show()
sns.heatmap(correlaton, annot=True, cmap='YlGnBu')
plt.show()
```

```python
q = np.append(np.arange(0, 1, 0.05), 1.0)
np.round(iris.quantile(q, numeric_only=True), 2)
```

```python
sns.pairplot(iris, hue='species', palette='rainbow', height=3)
plt.show()  
```

```python
sns.pairplot(iris, hue='species', palette='rainbow', height=5, x_vars=['sepal_length', 'sepal_width'], y_vars=['petal_length', 'petal_width'])
plt.show()
```

```python
sns.pairplot(iris, kind='reg')
plt.show()
```

```python
sns.pairplot(iris, height=3, vars=['sepal_length', 'sepal_width'])
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplots_adjust(hspace= 0.2, wspace= 0.2)
plt.subplot(2,2,1)
plt.title ('Sepal Length')
plt.plot(iris.sepal_length,'r')
plt.subplot(2,2,2)
plt.title ('Sepal Width')
plt.plot(iris.sepal_width, 'b')
plt.subplot(2,2,3)
plt.title ('Petal Length')
plt.plot(iris.petal_length, 'g')
plt.subplot(2,2,4)
plt.title ('Petal Width')
plt.plot(iris.petal_width,'y')
plt.show();
```

```python
plt.figure(figsize=(15,10))
sns.displot(iris.sepal_width)
plt.show()
```

```python
#Box Plot

plt.figure(figsize=(12,8))
sns.boxplot(data=iris, palette='Set2', orient='v')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='sepal_length', data=iris, linewidth=2)
plt.show()
```

```python
#Heatmap

import seaborn as sns

flight_data = sns.load_dataset('flights')
```

```python
flight_data.head()
```

```python
flight_data = flight_data.pivot(index='month', columns='year', values='passengers')
flight_data
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(flight_data, annot=True, fmt='d')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_data, cmap='rainbow')
```
