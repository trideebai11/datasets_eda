Data Cleaning:

```python
plt.figure(figsize=(12,7))
sns.boxplot(x='Pclass',y='Age', data=titanic, palette='winter', hue= 'Pclass', legend=False)
#creates a box plot using seaborn to visualize the distribution of ages (Age) across different passenger classes (Pclass) in the Titanic dataset
plt.show()
```

```python
def impute_age(cols): #here impute_age is a function that takes a single argument 'cols'
#The function expects a Series or list containing two values: Age and Pclass
    Age = cols.iloc[0]
    Pclass = cols.iloc[1]
    
    if pd.isnull(Age): #Checks if the Age value is missing (NaN)
        if Pclass == 1:
            return 37 #average age for Pclass 1
        elif Pclass == 2:
            return 29 #average age for Pclass 2
        else:
            return 24 #average age for Pclass 3
    else:
        return Age
```

```python
titanic['Age']= titanic[['Age','Pclass']].apply(impute_age,axis=1)
#applies the impute_age function to each row of the Titanic DataFrame, specifically to the 'Age' and 'Pclass' columns, and updates the 'Age' column with the imputed values
#axis=1 indicates that the function should be applied row-wise
```

```python
sns.heatmap(titanic.isnull(),yticklabels=False,cbar=False,cmap='viridis')
#creates a heatmap visualization using seaborn to represent missing data in the Titanic DataFrame, where missing values are highlighted in a different color
#yticklabels=False removes y-axis labels, cbar=False removes the color bar, and cmap='viridis' sets the color map for the heatmap
plt.show() #displays the plot in our output window
```
